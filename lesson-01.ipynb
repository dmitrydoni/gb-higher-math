#### Задание 4
Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

##### Решение

```python
import numpy as np
import matplotlib.pyplot as plt

# Display charts in Jupyter Notebook
%matplotlib inline

# Display inline plots as SVG
# %config InlineBackend.figure_formats = ['svg']

plt.style.use('fivethirtyeight')
```

```python
k1 = 1
k2 = 0.5

# Plot for y(k,x)=cos(k*x)
x = np.linspace(0, 10, 101)
plt.plot(x, np.cos(k1*x), color='green', label="k=1")
plt.plot(x, np.cos(k2*x), color='blue', label="k=0.5")

plt.xlabel('x')
plt.ylabel('cos(k*x)')

plt.legend(loc="upper right")
```
